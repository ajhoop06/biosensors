```python
from pdbfixer import PDBFixer
from openmm.app import PDBFile
```

```python
ID_list = ["0276", "0288", "0341", "0351", "0450", "0502", "0506", "0549", "0567", "0584", 
           "0602", "0647", "0684", "0736", "0748", "0765", "0776", "0782", "0811", "0822", 
           "0865", "0867", "0868", "0876", "0878", "0883", "0891", "0908", "0932", "0940",
           "0974", "0977", "1011", "1023", "1102", "1313", "1397", "1654", "1656", "1658", 
           "1671", "1679", "1709", "1710", "1737", "1758", "1784", "1785", "1792", "1835", 
           "1847", "1860", "1914", "1922", "1938", "1961", "1965"]

#seq_type = "binder"
seq_type = "nonbinder"
#seq_type = "neg_low_pkt"
#seq_type = "neg_fail_gate"

for i in range(len(ID_list)):
    ID = ID_list[i]

    if seq_type == "binder":
        #fixer = PDBFixer(filename=f"binders/seq{ID}_binder/protein_seq{ID}_binder.pdb")
        fixer = PDBFixer(filename=f"binders/pair_{ID}_binder/protein_pair{ID}.pdb")
    elif seq_type == "nonbinder":
        #fixer = PDBFixer(filename=f"nonbinders/seq{ID}_nb/protein_seq{ID}_nb.pdb")
        fixer = PDBFixer(filename=f"nonbinders/pair_{ID}_nb/protein_pair{ID}.pdb")
    elif seq_type == "neg_low_pkt":
        fixer = PDBFixer(filename=f"neg_low_pkt/pair_{ID}_low_pkt/protein_pair{ID}.pdb")
    elif seq_type == "neg_fail_gate":
        fixer = PDBFixer(filename=f"neg_fail_gate/pair_{ID}_fail_gate/protein_pair{ID}.pdb")

    # Detect problems
    fixer.findMissingResidues()
    fixer.findMissingAtoms()

    # Add missing heavy atoms
    fixer.addMissingAtoms()

    # Add hydrogens at desired pH
    fixer.addMissingHydrogens(pH=7.0)

    if seq_type == "binder":
        # with open(f"binders/seq{ID}_binder/protein_seq{ID}_binder_fixed_H.pdb", "w") as f:
        #     PDBFile.writeFile(fixer.topology, fixer.positions, f)
        with open(f"binders/pair_{ID}_binder/protein_pair{ID}_fixed_H.pdb", "w") as f:
            PDBFile.writeFile(fixer.topology, fixer.positions, f)
    elif seq_type == "nonbinder":
        # with open(f"nonbinders/seq{ID}_nb/protein_seq{ID}_nb_fixed_H.pdb", "w") as f:
        #     PDBFile.writeFile(fixer.topology, fixer.positions, f)
        with open(f"nonbinders/pair_{ID}_nb/protein_pair{ID}_fixed_H.pdb", "w") as f:
            PDBFile.writeFile(fixer.topology, fixer.positions, f)
    elif seq_type == "neg_low_pkt":
        with open(f"neg_low_pkt/pair_{ID}_low_pkt/protein_pair{ID}_fixed_H.pdb", "w") as f:
            PDBFile.writeFile(fixer.topology, fixer.positions, f)
    elif seq_type == "neg_fail_gate":
        with open(f"neg_fail_gate/pair_{ID}_fail_gate/protein_pair{ID}_fixed_H.pdb", "w") as f:
            PDBFile.writeFile(fixer.topology, fixer.positions, f)
```
